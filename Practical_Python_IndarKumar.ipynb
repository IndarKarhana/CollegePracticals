```python


#9) program to find factorial using recursion
def recur_factorial(n):  
   if n == 1:  
       return n  
   else:  
       return n*recur_factorial(n-1)  
# take input from the user  
num = int(input("Enter a number: "))  
# check is the number is negative  
if num < 0:  
   print("Sorry, factorial does not exist for negative numbers")  
elif num == 0:  
   print("The factorial of 0 is 1")  
else:  
   print("The factorial of",num,"is",recur_factorial(num)) 


#>>>OUTPUT
```

```python


#19) Program to fit Poisson distribution on a given data
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

data = np.random.normal(0, 0.5, 100000)
mean, var  = norm.fit(data)
x = np.linspace(-5,5,10000)
fitted_data = norm.pdf(x, mean, var)
plt.figure(figsize = (7,5))
plt.hist(data, density=True)
plt.plot(x,fitted_data,'r-')
plt.show()


#>>>OUTPUT
```

```python


#20) program to implement linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# generate random data-set
np.random.seed(10)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)

# Model initialization
regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(x, y)
# Predict
y_predicted = regression_model.predict(x)

# model evaluation
rmse = mean_squared_error(y, y_predicted)


# printing values
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
# plotting values
# data points
plt.figure(figsize = (10,7))
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color='g')
plt.title("Linear Regression")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()


#>>>OUTPUT
```
